***
**NumPy Practice:**
Below I practice using numpy arrays by calculating the max, mean, and index of the maximum value of an array. I also take the dot product of two arrays.
***

```python
import numpy as np

#Practicing with numpy arrays. 
array1 = np.array([1,2,3,4,5])
array2 = np.array([2.0,1.0,3.0,4.0,0])

print("The shape of the numpy array is " + str(array1.shape))
print("The type of the numpy array is " + str(array1.dtype))

#Using methods. 
print("The mean value of the array is " + str(array1.mean()))
print("The max value of the array is " + str(array1.max()))
print("The index of the max of value of the array is " + str(array1.argmax()))
print("The dot product is " + str(np.dot(array1,array2)))

#Note that numpy also has a random module. 
```

***
**MatPlotLib Practice:**
Below I use the matplotlib library to plot different functions.
***

```python
import matplotlib.pyplot as plt

#Get the x and y data. 
x = np.array([2,3,5,9,12])
y = [x[i]**3 for i in range(len(x))]
z = [x[i]**2 for i in range(len(x))]

#Adjust the axis labels.
plt.xlabel('Value of x')
plt.ylabel('Value of f(x)')

#Plot the data. 
plt.plot(x,y, label = "f(x) = x^3", color = "green")
plt.plot(x,z, label = "f(x) = x^2", color = "red")

#Add a title, a grid, and a legend.
plt.legend()
plt.title("Plot of x vs. f(x)")
plt.grid()

#Show the plot.
plt.show()
```
